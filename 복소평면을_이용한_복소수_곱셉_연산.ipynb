```python
import numpy as np
import matplotlib.pyplot as plt

def to_polar(z):
    return np.abs(z), np.angle(z)

def to_cartesian(r, theta):
    return complex(r * np.cos(theta), r * np.sin(theta))

def complex_to_str(z):
    real_part = f"{z.real:.2f}"
    imag_part = f"{abs(z.imag):.2f}i"
    if z.imag >= 0:
        return f"{real_part} + {imag_part}"
    else:
        return f"{real_part} - {imag_part}"

def plot_complex_operations(z1, z2, z_add, z_mul):
    fig, ax = plt.subplots(figsize=(6, 6))
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.grid(color='gray', linestyle='--', linewidth=0.5)

    ax.quiver(0, 0, z1.real, z1.imag, angles='xy', scale_units='xy', scale=1, color='r', label=f'Complex 1: {complex_to_str(z1)}')
    ax.quiver(0, 0, z2.real, z2.imag, angles='xy', scale_units='xy', scale=1, color='b', label=f'Complex 2: {complex_to_str(z2)}')
    ax.quiver(0, 0, z_add.real, z_add.imag, angles='xy', scale_units='xy', scale=1, color='orange', label=f'Addition Result: {complex_to_str(z_add)}')
    ax.quiver(0, 0, z_mul.real, z_mul.imag, angles='xy', scale_units='xy', scale=1, color='g', label=f'Multiplication Result: {complex_to_str(z_mul)}')

    ax.set_xlim(-20, 20)
    ax.set_ylim(-20, 20)
    ax.set_xlabel('Real Axis')
    ax.set_ylabel('Imaginary Axis')
    ax.legend()
    plt.show()

z1 = complex(input("Enter the first complex number (e.g., 3+2i): ").replace('i', 'j'))
z2 = complex(input("Enter the second complex number (e.g., 1-4i): ").replace('i', 'j'))

r1, theta1 = to_polar(z1)
r2, theta2 = to_polar(z2)
r_mul, theta_mul = r1 * r2, theta1 + theta2

z_add = z1 + z2
z_mul = to_cartesian(r_mul, theta_mul)

plot_complex_operations(z1, z2, z_add, z_mul)

print(f"Polar form of Complex 1: Magnitude = {r1:.2f}, Angle = {np.degrees(theta1):.2f} degrees")
print(f"Polar form of Complex 2: Magnitude = {r2:.2f}, Angle = {np.degrees(theta2):.2f} degrees")
print(f"Polar form of Multiplication Result: Magnitude = {r_mul:.2f}, Angle = {np.degrees(theta_mul):.2f} degrees")
print(f"Addition Result: {complex_to_str(z_add)}")
print(f"Multiplication Result: {complex_to_str(z_mul)}")
```
